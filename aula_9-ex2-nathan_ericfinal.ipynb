```python
def graf_lognormal(sigma,cor1,cor2):
    x = pd.Series(range(1,1001))/100
    lognormal = stats.lognorm.pdf(x,sigma)
    acumulativa = stats.lognorm.cdf(x,sigma)
    xi = stats.lognorm.ppf(0.001,sigma)
    xf = stats.lognorm.ppf(0.999,sigma)
    plt.plot(x,lognormal,cor1,label='Lognormal')
    plt.plot(x,acumulativa,cor2, label='Acumulativa')
    plt.legend()
    #plt.axis([xi,xf])
    plt.show()
```

```python
graf_lognormal(0.4,'b','r')
graf_lognormal(0.6,'b','r')
graf_lognormal(0.8,'b','r')
graf_lognormal(1,'b','r')
graf_lognormal(1.5,'b','r')
graf_lognormal(5,'b','r')
```
